```python
import pandas as pd

# Load datasets (correct path)
train = pd.read_csv('data_row/train.csv')
features = pd.read_csv('data_row/features.csv')
stores = pd.read_csv('data_row/stores.csv')

# Merge datasets
df = train.merge(features, on=['Store','Date','IsHoliday'], how='left')
df = df.merge(stores, on='Store', how='left')

# Convert date column
df['Date'] = pd.to_datetime(df['Date'])

# Fill missing markdowns with 0
markdown_cols = ['MarkDown1','MarkDown2','MarkDown3','MarkDown4','MarkDown5']
df[markdown_cols] = df[markdown_cols].fillna(0)

# Forward fill CPI & Unemployment
df['CPI'] = df['CPI'].fillna(method='ffill')
df['Unemployment'] = df['Unemployment'].fillna(method='ffill')

# Create time features
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Week'] = df['Date'].dt.isocalendar().week

# Save cleaned dataset
df.to_csv('data_clean/walmart_clean.csv', index=False)

df.head()
```
