```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('data_clean/walmart_clean.csv', parse_dates=['Date'])

# --- 1. Store-level sales trend ---
store1 = df[df['Store']==1].groupby('Date')['Weekly_Sales'].sum()

plt.figure(figsize=(12,5))
plt.plot(store1.index, store1.values)
plt.title('Weekly Sales Trend - Store 1')
plt.xlabel('Date')
plt.ylabel('Weekly Sales')
plt.show()

# --- 2. Department-level trend for Store 1 ---
dept1 = df[(df['Store']==1)&(df['Dept']==1)].sort_values('Date')

plt.figure(figsize=(12,5))
plt.plot(dept1['Date'], dept1['Weekly_Sales'])
plt.title('Dept 1 Sales - Store 1')
plt.xlabel('Date')
plt.ylabel('Weekly Sales')
plt.show()

# --- 3. Holiday vs non-holiday comparison ---
holiday = df[df['IsHoliday']==True]['Weekly_Sales'].mean()
nonholiday = df[df['IsHoliday']==False]['Weekly_Sales'].mean()

print("Holiday Avg Sales:", holiday)
print("Non-Holiday Avg Sales:", nonholiday)
```
